```python
import pandas as pd
```

```python
df1=pd.read_csv("F:/denoised/split/I/mlp/train_mlp.csv")
```

```python
df2=pd.read_csv("F:/denoised/split/I/mlp/test_mlp.csv")
```

```python
df1=df1.drop([ 'Unnamed: 0'], axis=1)
df2=df2.drop([ 'Unnamed: 0'], axis=1)
```

```python
df1=df1.drop([ 'FileName'], axis=1)
df1=df1.drop([ 'Beat'], axis=1)
df1=df1.drop([ 'PatientAge'], axis=1)
df1=df1.drop([ 'Gender'], axis=1)
```

```python
df2=df2.drop([ 'FileName'], axis=1)
df2=df2.drop([ 'Beat'], axis=1)
df2=df2.drop([ 'PatientAge'], axis=1)
df2=df2.drop([ 'Gender'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder 
```

```python
label = LabelEncoder()
df1['label'] = label.fit_transform(df1['Rhythm'])
df1.head()
```

```python
df2['label'] = label.fit_transform(df2['Rhythm'])
df2.head()
```

```python
train_X = df1[['VentricularRate','AtrialRate','QRSDuration','QTInterval','QTCorrected','RAxis','TAxis','QRSCount','QOnset','QOffset','TOffset']]
train_y = df1['label']
```

```python
test_X = df2[['VentricularRate','AtrialRate','QRSDuration','QTInterval','QTCorrected','RAxis','TAxis','QRSCount','QOnset','QOffset','TOffset']]
test_y = df2['label']
```

```python
scaler = StandardScaler()
```

```python
train_X = scaler.fit_transform(train_X)
```

```python
train_X = pd.DataFrame(data = train_X, columns = ['VentricularRate','AtrialRate','QRSDuration','QTInterval','QTCorrected','RAxis','TAxis','QRSCount','QOnset','QOffset','TOffset'])
```

```python
#scaled_X['label'] = y.values
train_X
```

```python
test_X = scaler.fit_transform(test_X)
```

```python
test_X = pd.DataFrame(data = test_X, columns = ['VentricularRate','AtrialRate','QRSDuration','QTInterval','QTCorrected','RAxis','TAxis','QRSCount','QOnset','QOffset','TOffset'])
```
